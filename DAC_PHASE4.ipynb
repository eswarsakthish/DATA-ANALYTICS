```python
# Display basic information about the dataset
print(data.info())

# Explore the first few rows of the dataset
print(data.head())

# Question 1: Calculate the average age of the respondents
average_age = data['Age'].mean()
print(f"Average Age: {average_age:.2f} years")

# Question 2: Count the number of respondents by gender
gender_counts = data['Gender'].value_counts()
print("Gender Counts:")
print(gender_counts)

# Question 3: Count the number of respondents from each country
country_counts = data['Country'].value_counts()
print("Country Counts:")
print(country_counts)

# Question 4: Calculate the percentage of respondents who have a family history of mental health issues
family_history_percentage = (data['family_history'].value_counts(normalize=True) * 100)['Yes']
print(f"Percentage with Family History: {family_history_percentage:.2f}%")

# Question 5: Count the number of respondents who receive mental health treatment
treatment_counts = data['treatment'].value_counts()
print("Treatment Counts:")
print(treatment_counts)
```
