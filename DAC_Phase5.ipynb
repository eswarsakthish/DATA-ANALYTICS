```python
#importing the necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
#importing the dataset
df = pd.read_csv(r"survey.csv")
```

```python
df
```

```python
#droping unwanted columns
df = df.drop(['Timestamp','comments'],axis = 1)
```

```python
df.info()
```

```python
#droping null values from the dataset
df = df.dropna()
```

```python
df
```

```python
sns.countplot(data = df, x= "family_history", hue= 'mental_health_consequence') 
# people with no family history of mental health issues tend to have lower mental issues
```

```python
sns.countplot(data = df, x= "family_history", hue= 'phys_health_consequence')
# people with family history of mental health issues tend to have slightly higher chance of having physical issues
```

```python
sns.countplot(data = df, x= "remote_work", hue= 'phys_health_consequence')
#people who work remotely tend to have no physical issues
```

```python
sns.countplot(data = df, x= "remote_work", hue= 'mental_health_consequence')
#people who work remotely have low chance for mental issues
```

```python
sns.histplot(data = df, x = 'Age', hue = 'mental_health_consequence',bins = 50 )
#people in the age group of 20 - 50 tend to have mental issues
```

```python
sns.countplot(data = df, x= 'self_employed', hue = "mental_health_consequence")
#people who are self employed have low chance of having mental health issues
```

```python
sns.countplot(data = df, x = 'treatment', hue = 'mental_health_consequence')
#people who have undergone treatment for mental health issues have lower chance of having mental health issues
```

```python
sns.countplot(data = df, x = 'tech_company', hue = 'mental_health_consequence')
#people who work in tech companies have more chance of having mental health issues
```

```python
sns.countplot(data = df, x = 'seek_help', hue = 'mental_health_consequence')
#people who seek help for mental health issues have more chance of improvement than those who don't
```
